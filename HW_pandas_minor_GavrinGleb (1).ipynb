```python
#16
#перевод баллов в оценки
def GetGrade(average_mark):
  mark = ''
  if 0 <= average_mark <50:
    mark = 'F'
  elif 50 <= average_mark <60:
    mark = 'E'
  elif 60 <= average_mark <70:
    mark = 'D'
  elif 70 <= average_mark <80:
    mark = 'C'
  elif 80 <= average_mark <90:
    mark = 'B'
  elif 90 <= average_mark:
    mark = 'A'
  return mark
df['Grade'] = df[df.columns[-7:-4]].mean(axis=1).apply(lambda x: GetGrade(x))
df
```
